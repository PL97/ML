```python
plt.figure(figsize=(16, 6))

from sklearn import datasets
iris = datasets.load_iris()
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
plt.title("line plot for iris")
sns.lineplot(data=data1.iloc[:, 0], label = "sepal length")
sns.lineplot(data=data1.iloc[:, 1], label = "sepal width")
plt.xlabel("index")
plt.ylabel("length")
print(data1.columns)
```

```python
data1.head()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(y = data1.index, x = data1.iloc[:, 0])
```

```python
plt.figure(figsize=(10, 30))
sns.heatmap(data = data1, annot=True)
```

```python
print(data1.columns)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data1["sepal length (cm)"], y = data1["sepal width (cm)"], hue = data1["target"].astype("int"))
# sns.regplot(x=data1.iloc[:, 0], y = data1.iloc[:, 1])
```

# use name of a attribute to refer data

```python
sns.set_style("dark")
setosa = data1.loc[data1["target"] == 0]
versicolor = data1.loc[data1["target"] == 1]
virginica = data1.loc[data1["target"] == 2]
dist1 = sns.distplot(a = setosa['petal length (cm)'], kde = False)
dist2 = sns.distplot(a = versicolor['petal length (cm)'], kde = False)
dist3 = sns.distplot(a = virginica['petal length (cm)'], kde = False)
plt.title("Histogram of Petal Lengths, by Species")
# plt.legend((dist1, dist2, dist3), ('label1', 'label2', 'label3'), loc='upper right')
```
