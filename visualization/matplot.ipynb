```python
plt.plot(np.random.randn(30), color = 'k', linestyle = 'dashed', marker = 'o')
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label = 'Default')
```

```python
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.plot(data, 'k--', label='steps-post')
plt.legend(loc='best')
plt.xlim([0, 35])
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
ax[0, 0].scatter(x[0], x[1], label='one')
ticks = ax[0, 0].set_xticks([0, 2, 4])
ax[0, 0].set_title('scatter')
ax[0, 0].set_xlabel("X")
ax[0, 0].set_ylabel("Y")
ax[0, 0].legend(loc = 'best')

ax[0, 1].plot(x[0])
ticks = ax[0, 0].set_xticklabels(['one', 'two', 'three', 'four'], rotation=30, fontsize = 'small')
ax[0, 1].set_title('plot')
ax[0, 1].text(2, 0, "test", fontsize = 10)
```
