sigmod function

$$y = \frac{1}{1 + e^{-w^Tx + b}}$$

$$ln\frac{y}{1-y} = w^Tx + b$$

let $p_0 = y$ and $p_1 = 1-y$

$$p(y = 0|x) = \frac{e^{w^tx + b}}{1 + e^{w^tx + b}}$$

$$p(y = 1|x) = \frac{1}{1 + e^{w^tx + b}}$$

according to maximum likehood method

$$l(w, b) = \sum^{m}_{i=1}ln p(y_i|x_i;w,b)$$

which equals to

$$l(w, b) = \sum^{m}_{i=1} y ln p_1(y_i|x_i;w,b) + (y-1) ln p_0(y_i|x_i;w,b)$$

$$\frac{\partial l(\beta)}{\partial \beta} = \sum^m_{i=1} -y_ix_i + \frac{e^{w^tx_i + b}}{1 + e^{w^tx_i + b}} x_i$$

```python
x1 = np.array(range(0, 100))
x2 = np.array(range(200, 300))

y1 = np.random.normal(75, 60, 100)
y2 = np.random.normal(50, 60, 100)
b = np.ones(100)

X1 = np.stack((x1, y1, b), axis=1)
X2 = np.stack((x2, y2, b), axis=1)

print(X1.shape)
X = np.concatenate((X1, X2))
c1 = np.ones(100)
c2 = np.zeros(100)
y = np.concatenate((c1, c2))
print(X.shape, y.shape)
```

```python
import seaborn as sns
sns.scatterplot(x1, y1)
sns.scatterplot(x2, y2)

import seaborn as sns
import matplotlib.pyplot as plt

def plot(w):
    plt.figure()
    sns.scatterplot(x1, y1)
    sns.scatterplot(x2, y2)

    xx = list(range(0, 300))
    print(-w[0]/w[1])
    yy = [-w[0]/w[1] * xxx  - w[2]/w[1] for xxx in xx]
    sns.lineplot(x = xx, y = yy)
    plt.show()
```

```python
def getP(x, w):
    if np.dot(x, w.transpose()) > 200:
        return 1
    return np.exp(np.dot(x, w.transpose()))/(np.exp(np.dot(x, w.transpose())) + 1)

def lpr(x, y):
    w = np.array([1, 2, 3])
    lamda = 0.01
    gradient = 0
    gradient2 = 0
    for iteration in range(5000):
        for i in range(x.shape[0]):
            gradient = gradient - x[i, :] * y[i] 
            gradient = gradient + x[i, :] * getP(x[i, :], w)
            
            gradient2 = gradient2 +  np.dot(x[i, :].transpose(), x[i, :]) * getP(x[i, :], w)*(1-getP(x[i, :], w))
            
        pre_w = w
        w = w - lamda * (1/gradient2) * gradient
        
        change = sum(np.square(w - pre_w))
#         if(change < 0.000001):
#             break
        if iteration % 100 == 0:
            print("{}th iteration:{}".format(iteration, change))
            print(w)
            plot(w)
    return w
#     print(ret)

w = lpr(X, y)
```
